```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("../notebook/data/HR_comma_sep.csv")
```

```python
print(df.head())
print(df.info())
print(df.describe())
```

```python
print(df.isnull().sum())
```

```python
sns.countplot(x='left', data=df)
plt.title("Employee Attrition (Left the company)")
plt.show()
```

```python
sns.boxplot(x='left', y='satisfaction_level', data=df)
plt.show()
```

```python
df_encoded = pd.get_dummies(df, columns=['Department', 'salary'], drop_first=True)
```

```python
df_encoded.to_csv("HR_encoded.csv", index=False)
```
